```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Loading Data
 
To look for connections between GDP and life expectancy load the datasets into DataFrames so that they can be visualized.

```python
df = pd.read_csv('all_data.csv')
df.head()
```

# Explore Data 

## Has life expectancy increased over time in the six nations?
First we need to see which countries are there

```python
print(df.Country.unique())
```

Second we need to see how many years

```python
print(df.Year.unique())
```

## Rename column name
the name is too long(Life expectancy at birth). i want to work with more efficient name 

```python
df = df.rename({"Life expectancy at birth (years)":"LEAB"}, axis = "columns")
df.head()
```

## Creating a Line plot 
every country over the years increased expectancy of life, Zimbabwe has the most growth since 2004

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year , y=df.LEAB, hue= df.Country)
plt.legend(loc= 'best' ,bbox_to_anchor=(1, 1))

```

## Has GDP increased over time in the six nations?
most of the countries increased GPD over time, Zimbabwe is the only country that haven't increased their GDP.
China is the country who grew the most over the years 

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year , y=df.GDP, hue= df.Country)
plt.legend(loc= 'best' ,bbox_to_anchor=(1, 1))
```

## Is there a correlation between GDP and life expectancy of a country?
Yes, as you can see the GDP growing as long as the life expectacy growing.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df.LEAB, y=df.GDP, hue=df.Country)
plt.legend(loc= 'best' ,bbox_to_anchor=(1, 1))
```

## What is the average life expectancy in these nations?
most of the countries are mid-to-high 70s except for zimbabwe which is 50

```python
df_mean = df.drop("Year", axis = 1).groupby("Country").mean().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x="Country", y="LEAB", data=df_mean)
plt.xticks(rotation = 15)
plt.xlabel("Life expectancy at birth (years)");
```

## What is the distribution of that life expectancy? 
The distribution of life expectancy in the data is very left skewed where most of the values are on the right-hand side.

```python
plt.figure(figsize=(8,6))
plt.hist(df.LEAB,25)
plt.xlabel("Life expectancy at birth (years)");
```
